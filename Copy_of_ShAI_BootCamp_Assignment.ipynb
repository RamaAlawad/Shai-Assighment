# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



```

# Load the data set


```python
df =pd.read_csv('Salaries.csv')
df.head()
```

# Basic Data Exploration

```python
#number of columns and rows
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
#Data Types
print("\nData types of each column:")
print(df.dtypes)
# Check for missing values in each column
print("\nMissing values in each column:")
print(df.isnull().sum())
```

# Descriptive Statistics

```python
print("\nDescriptive statistics for numerical columns:")
print(df.describe())
```

# Data Cleaning

```python
# Handle missing data by filling NaN values with the mean for numerical columns
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
```

```python
print("\nDescriptive statistics for numerical columns:")
print(df.describe())
```

#Basic Data Visualization:





```python
# Histogram to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Use a pie chart to represent the proportion of employees in different departments
plt.figure(figsize=(10, 6))
df['JobTitle'].value_counts().head(10).plot.pie(autopct='%1.1f%%')
plt.title('Top 10 Job Titles Distribution')
plt.show()
```

#Grouped Analysis


```python
job_title_groups = df.groupby('JobTitle')['TotalPay'].agg(['mean', 'median', 'count']).sort_values(by='mean', ascending=False)
print("\nGrouped Analysis by Job Title:")
print(job_title_groups.head(10))
```

# Simple Correlation Analysis

```python
# Identify correlation between 'TotalPay' and 'BasePay'
correlation = df['TotalPay'].corr(df['BasePay'])
print("\nCorrelation between TotalPay and BasePay:", correlation)

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BasePay', y='TotalPay', data=df)
plt.title('Scatter Plot: TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```
